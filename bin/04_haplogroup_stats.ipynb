```python
import pandas as pd
import os
from importlib import reload
import haplogroup_analyzer as ha
import pandas as pd
reload(ha)
PATH = os.path.dirname(os.getcwd())
```

```python
haplo_stats = pd.read_csv(os.path.join(PATH, 'data', 'haplogroups', 'all_haplogroups.csv'), index_col = 0)
# Remove " from all values
haplo_stats = haplo_stats.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)
# Do the same for column names
haplo_stats.columns = haplo_stats.columns.str.replace('"', '')
# Grab the first string element of each Haplogroup as a baseHaplo
haplo_stats['baseHaplo'] = haplo_stats['Haplogroup'].apply(lambda x:x[0])
haplo_stats['SampleID'] = haplo_stats['SampleID'].apply(lambda x: x.split('.')[0])
haplo_stats.to_csv(os.path.join(PATH, 'data', 'haplogroups', 'all_haplogroups_cleaned.csv'))

display(haplo_stats.head())
```
